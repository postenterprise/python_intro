```python
# import numpy library
import numpy as np

# import graphics plotting library
import matplotlib.pyplot as plt

# import more graphics functions
import scipy.ndimage as ndi

# make graphics plotting inline
%matplotlib inline  
```

```python
a = np.random.power(1, size=(100, 100))
```

```python
plt.imshow(a, cmap='copper')
plt.colorbar()
```

```python
b = ndi.filters.gaussian_filter(a,1)
```

```python
plt.imshow(b, cmap='copper')
plt.colorbar()
```
